```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.sparse import diags
from scipy.interpolate import interp1d
from aux_functions import events_90, events_90_flux
```

```python
int_frac = 0.95
```

```python
## indices 0 through 7 are time and delta mag in u, g, ,r, i, z, and y respectively
data = np.loadtxt('lc_examples/0.dat')
time = data[:,0]
mag  = data[:,3]

## this function takes 3 arguments: x, y, and a threshold to define the event
time_start, time_end, mag_start, mag_end, _ = events_90_flux(time, mag, 0.3, int_frac)

## plot
sns.set_style("darkgrid")
sns.set(font_scale = 2)

fig, (ax1, ax2) = plt.subplots(1,2 ,figsize=(24,12))

ax1.title.set_text('using flux')
ax1.invert_yaxis()
ax1.plot(time, mag, 'k')
ax1.scatter(time_start, mag_start, marker='+', color='green', label='start of event', s=500)
ax1.scatter(time_end, mag_end, marker='x', color='red', label='end of event', s=500)
ax1.set_xlabel('time [days]')
ax1.set_ylabel('$\Delta$ mag')
ax1.legend()

## this function takes 3 arguments: x, y, and a threshold to define the event
time_start, time_end, mag_start, mag_end, _ = events_90(time, mag, 0.3, int_frac)

ax2.title.set_text('using mag')
ax2.invert_yaxis()
ax2.plot(time, mag, 'k')
ax2.scatter(time_start, mag_start, marker='+', color='green', label='start of event', s=500)
ax2.scatter(time_end, mag_end, marker='x', color='red', label='end of event', s=500)
ax2.set_xlabel('time [days]')
ax2.set_ylabel('$\Delta$ mag')
ax2.legend();
```
