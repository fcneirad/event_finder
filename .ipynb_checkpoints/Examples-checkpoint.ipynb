```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.sparse import diags
from scipy.interpolate import interp1d
from aux_functions import events_90, events_90_flux
```

```python
## indices 0 through 7 are time and delta mag in u, g, ,r, i, z, and y respectively
data = np.loadtxt('lc_examples/0.dat')
time = data[:,0]
mag  = data[:,3]

## this function takes 3 arguments: x, y, and a threshold to define the event
time_start, time_end, mag_start, mag_end, _ = events_90_flux(time, mag, 0.3)

## plot
sns.set_style("darkgrid")
sns.set(font_scale = 2)

plt.figure(figsize=(12,12))
plt.title('Example 0')
plt.gca().invert_yaxis()
plt.plot(time, mag, 'k')
plt.scatter(time_start, mag_start, marker='+', color='green', label='start of event', s=500)
plt.scatter(time_end, mag_end, marker='x', color='red', label='end of event', s=500)
plt.xlabel('time [days]')
plt.ylabel('$\Delta$ mag')

plt.legend();
```
